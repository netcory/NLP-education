```python
def scroll_down(driver):
    driver.execute_script('window.scrollTo(0, 19431049)')
    time.sleep(1)
    
n = 5
i = 0
while i < n: # 5번 스크롤 내리기 위해
    scroll_down(driver)
    i += 1

# 스크롤 19431049만큼 내리고 1초 쉬고, 다시 스크롤 내리고 1초 쉬고 다섯번 반복
```

```python
url_list = []
title_list = []

articles = '.api_txt_lines.total_tit'
article_raw = driver.find_elements_by_css_selector(articles)
article_raw
```

```python
article_raw[0].get_attribute('href') # url 주소 가져와주는 메소드 get_attribute
```

```python
for article in article_raw:
    url = article.get_attribute('href')
    url_list.append(url)
time.sleep(1)

for article in article_raw:
    title = article.text
    title_list.append(title)
    
    print(title)
    
print("")
print('url갯수: ', len(url_list))
print('title갯수: ', len(title_list))
```
